## step 1: get a list of all the programs
this list will have the dashes included so it can be used later for a looped scrape of all required courses pages for the pgorams.



```python
#importing all the libraries we will use.

import requests
from bs4 import BeautifulSoup

import pandas as pd

#using the wctc A-Z list of programs url

url = 'https://www.wctc.edu/academics/programs/programs-a-z.php'
    
    
response = requests.get(url)

if response.status_code == 200:
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the body tag
    body = soup.body
    
    if body:
        # Find all a tags within the body
        a_tags = body.find_all('a')
        
        # Extract href attributes and save to a list
        links = [a.get('href') for a in a_tags if a.get('href')]
        
        # Print the links
        for link in links:
            print(link)
    else:
        print("No body tag found in the HTML document.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

# The 'links' list now contains all the links
#print("Extracted links:", links)

linksdf = pd.DataFrame(links, columns=['Filtered Text'])

# Save the DataFrame to a .csv file for later use (also so I dont need to constantly scrape wctc. cuz im sure they dont enjoy it)
#df.to_csv('list of all programs.csv', index=False)

# Output to confirm the operation (print the DataFrame)
linksdf
```

```python
import re
import pandas as pd
def extract_text(link):
    # Use regex to find the text following 'programs/'
    match = re.search(r'programs?/(.+?)(?:/|$)', link)
    if match:
        return match.group(1)
    return None

# Create a new list with the extracted text
extracted_texts = []

extracted_texts
# Iterate through each link and append the extracted text if it exists
for link in links:
    text = extract_text(link)
    if text:  # Only append if text is not None or empty
        extracted_texts.append(text)
extracted_texts

programnames = pd.DataFrame(extracted_texts, columns=['Filtered Text'])

# Save the DataFrame to a .csv file for later use (also so I dont need to constantly scrape wctc. cuz im sure they dont enjoy it)
programnames.to_csv('list of all programs.csv', index=False)

# Output to confirm the operation (print the DataFrame)
print(programnames)
```

## Data clean up
we now have a pandas dataframe.
However, there are some messy parts about it.
for one, the Approved subsitutes are being put into a row.
My hypothesis is that the Approved subsitutes appear under the course it applies to.
I am going to first manually search wctc to make sure this is true.

I did and my hypothesis is correct.

I have two choices.

1. just drop the rows for approved subs and assume students will do things the suggested way 

2. create a new column named "Approved Substitute"  containing the course code(s) of the appproved sub.

I will go with option 2 because I like keeping data.

Other Notes:
The first row just contains index info. we can drop this row from the start.





```python
#first lets drop that first row and last row because these are unnecessary. 

df.drop(index=[df.index[0], df.index[-1]], inplace=True)

#now lets add a new column with the default value 'No Substitute'

df['Approved Substitute'] = 'No Approved Substitutes'

# Identify the rows starting with 'Approved Substitute:' or 'Approved Substitutes:'
mask = df['Code'].str.startswith('Approved Substitute:') | df['Code'].str.startswith('Approved Substitutes:')

# Extract the text after 'Approved Substitute:' or 'Approved Substitutes:'
df.loc[mask, 'Approved Substitute'] = df.loc[mask, 'Code'].str.replace(r'Approved Substitutes?:', '', regex=True).str.strip()


# Shift the 'Approved Substitute' column up by 1 row
df['Approved Substitute'] = df['Approved Substitute'].shift(-1)

# Drop the rows that start with 'Approved Substitute:' or 'Approved Substitutes:'
df = df[~mask]

df = df.dropna(subset=['Credits'])

# Reset the index if necessary
df = df.reset_index(drop=True)

df['Completed'] = False


RequiredCourses = df

RequiredCourses
```

## Lets Automate and loop this.

```python
from requests.exceptions import HTTPError

def ScrapeCourses(program):
    url = f'https://catalog.wctc.edu/programs/{program}/#requiredcoursestext'
    try:
        tables = pd.read_html(url)
        df = tables[0]
        df.drop(index=[df.index[0], df.index[-1]], inplace=True)
        df['Approved Substitute'] = 'No Approved Substitutes'
        mask = df['Code'].str.startswith('Approved Substitute:') | df['Code'].str.startswith('Approved Substitutes:')
        df.loc[mask, 'Approved Substitute'] = df.loc[mask, 'Code'].str.replace(r'Approved Substitutes?:', '', regex=True).str.strip()
        df['Approved Substitute'] = df['Approved Substitute'].shift(-1)
        df = df[~mask]
        df = df.dropna(subset=['Credits'])
        df = df.reset_index(drop=True)
        df['Completed'] = False
        RequiredCourses = df
        file_name = f'{program} Required Courses.xlsx'
        writer = pd.ExcelWriter(file_name, engine='xlsxwriter')

# Write the DataFrame to the Excel file
        RequiredCourses.to_excel(writer, sheet_name='ProgramProgress', index=False, header=RequiredCourses.columns)

# Access the xlsxwriter workbook and worksheet objects for further customization
        workbook = writer.book
        worksheet = writer.sheets['ProgramProgress']

# Add a table style to the data range
        num_rows = len(df)  # Number of rows in the DataFrame
        num_cols = len(df.columns)  # Number of columns in the DataFrame
        cell_range = f'A1:E{num_rows+1}'  # Adjust based on your columns

# Add table style
        worksheet.add_table(cell_range, {'style': 'Table Style Medium 9'})

# Apply conditional formatting to 'Completed' column
        green_format = workbook.add_format({'bg_color': '#C6EFCE', 'font_color': '#006100'})
        red_format = workbook.add_format({'bg_color': '#FFC7CE', 'font_color': '#9C0006'})

        worksheet.conditional_format(f'E2:E{num_rows+1}', {'type': 'cell',
                                                           'criteria': 'equal to',
                                                           'value': True,
                                                           'format': green_format})

        worksheet.conditional_format(f'E2:E{num_rows+1}', {'type': 'cell',
                                                           'criteria': 'equal to',
                                                           'value': False,
                                                           'format': red_format})

# Write totals to the Excel file
        bold = workbook.add_format({'bold': True})
        worksheet.write(num_rows + 1, 1, 'Total Credits Required:', bold)
        worksheet.write(0, 0, 'Code', bold)
        worksheet.write(0, 1, 'Title', bold)
        worksheet.write(0, 2, 'Credits', bold)
        worksheet.write(0, 3, 'Approved Substitute', bold)
        worksheet.write(0, 4, 'Completed', bold)
        worksheet.write(num_rows + 1, 2, f'=SUM(C2:C{num_rows+1})')
        worksheet.write(num_rows + 2, 1, 'Total Credits Completed:', bold)
        worksheet.write(num_rows + 2, 2, f'=SUMIFS(C2:C{num_rows+1}, E2:E{num_rows+1}, TRUE)')

#im also going to adjust the width of the columns to be more appropriate
        worksheet.set_column('A:A', 8)  # Code will be 8
        worksheet.set_column('B:B', 29)  #Title 29
        worksheet.set_column('C:C', 9)  # Credits 9
        worksheet.set_column('D:D', 38)  #  38
        worksheet.set_column('E:E', 9)  #  9

# Save the Excel file
        writer.close()
    except HTTPError as http_err:
        print(f'HTTP error occurred while fetching {url}: {http_err}')
        errorlog.append(program)
        # Handle specific errors if needed
    except Exception as err:
        print(f'Other error occurred while fetching {url}: {err}')
        errorlog.append(program)

        # Handle other exceptions if needed
    else:
        print(f'Successfully fetched data from {url}')

        
    
```
